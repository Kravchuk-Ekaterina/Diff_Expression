```python
import pandas as pd
```

Let's look at the GO Slim Mapper results

```python
slim_mapper = pd.read_csv('../data/mapper_genes_5768_slimTab.txt', sep='\t')
slim_mapper.head()
```

Let's observe changes in the expression of our 50 genes:

```python
genes = pd.read_csv('../data/genes_50.txt', sep='\t')[['id', 'log2FoldChange', 'lfcSE', 'padj']]
genes.id = genes.id.str.split('-', 1).str[-1]
genes['TERM'] = ''
genes.head()
```

Merging the results:

```python
for i, r in slim_mapper.iterrows():
    for g in r.ANNOTATED_GENES.split(', '):
        genes.loc[genes.id == g, 'TERM'] = genes.loc[genes.id == g, 'TERM'] + r.TERM + ', '
```

The genes with positive changes in expression level:

```python
from IPython.display import display, HTML

display(HTML(genes[genes['log2FoldChange'] > 0].to_html()))
```

The genes with negative changes

```python
display(HTML(genes[genes['log2FoldChange'] < 0].to_html()))
```
